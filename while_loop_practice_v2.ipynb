## Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
#write answer here
n = int(input("Enter a number: "))

total = 0

# Add numbers in loop
for i in range(1, n + 1):
    total += i

print("Sum from 1 to", n, "is:", total)
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here

var= int(input())

new = var

res=0

while var>0:
  temp = var % 10
  res = res*10 + temp
  var = var // 10

print(res)


if new==res:
  print("Palindrome")
else:
  print("Not Palindrome")


```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
num = int(input("Enter a number: "))

count = 0

# Handle negative numbers
num = abs(num)

# Divide number by 10 repeatedly
while num > 0:
    num = num // 10
    count += 1

print("Total digits:", count)
```

*Q7*. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here

num = int(input("Enter a number: "))

original = num
reverse = 0

# Reverse number
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

# Compare with original
if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")

```
